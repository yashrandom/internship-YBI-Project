```python
#importing pandas and msno
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Download diabetes_dataset_with_missing_values.csv")
```

```python
#Fill missing entries for Pregnancies
df['Pregnancies'] = df['Pregnancies'].fillna(df['Pregnancies'].median())
```

```python
#Fill missing entries for Glucose
df['Glucose'] = df['Glucose'].fillna(df['Glucose'].median())
```

```python
#Fill missing entries for BloodPressure
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].median())
```

```python
#Fill missing entries for SkinThickness
df['SkinThickness'] = df['SkinThickness'].fillna(df['SkinThickness'].median())
```

```python
#Fill missing entries for Insulin
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].median())
```

```python
#Fill missing entries for BMI
df['BMI'] = df['BMI'].fillna(df['BMI'].median())
```

```python
#Fill missing entries for DiabetesPedigreeFunction
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].fillna(df['DiabetesPedigreeFunction'].median())
```

```python
#Fill missing values for Age
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.info()
```

```python
df.columns
```

```python
# Define y and X
X = df.drop(['Outcome'],axis=1)
y = df['Outcome']
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Select Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1200)
```

```python
#Train Model
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#Prediction
y_pred = model.predict(X_test)
```

```python
#Evaluate
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
